```python
class Module:
    def __init__(self, name, c_list=[], a_list=[], d_list=[], group="action"):
        self.name = name #str型
        self.c_list = c_list #リスト型
        self.a_list = a_list #リスト型
        self.d_list = d_list #リスト型
        self.activation_level = 0 #int型
        self.threshold = 100 #int型のパラメータ（モジュールごとに変えなくていいよね？？）
        self.group = group
        
    def act(self):
            print("activated " + self.name + "!")
            #sys.stdout.write(self.name)
            activate = True
            return activate
```

```python
import copy
class STRIPS:
    def __init__(self):
        # current_stateは，環境リストをコピーしたものとする．
        self.env = AnaEnv("drink_water")
        self.current_state = self.env.get_current_state()
        self.current_step = 0
        self.influence_from_goals = 100
        self.influence_from_state = 10
        self.influence_from_achieved_goals = 60
        self.cup_intensity = 30
        self.bottle_intensity = 1
        self.mean_activation_level = 15
        self.drink = Module("drink", c_list=["cup_at_mouth"], a_list=["relieve_thirst"], d_list=[""], group="action")
        self.pick_up_A = Module("pick_up_A", c_list=["clear_A", "on_table_A","arm_empty"], a_list=["holding_A"], d_list=["on_table_A","arm_empty"], group="action")
        self.pick_up_B = Module("pick_up_B", c_list=["clear_B", "on_table_B","arm_empty"], a_list=["holding_B"], d_list=["on_table_B","arm_empty"], group="action")
        self.pick_up_C = Module("pick_up_C", c_list=["clear_C", "on_table_C","arm_empty"], a_list=["holding_C"], d_list=["on_table_C","arm_empty"], group="action")
        self.stack_A_on_B = Module("stack_A_on_B", c_list=["clear_B", "holding_A"], a_list=["arm_empty", "A_on_B"], d_list=["holding_A","clear_B"], group="action")
        self.stack_A_on_C = Module("stack_A_on_C", c_list=["clear_C", "holding_A"], a_list=["arm_empty", "A_on_C"], d_list=["holding_A","clear_C"], group="action")
        self.stack_B_on_A = Module("stack_B_on_A", c_list=["clear_A", "holding_B"], a_list=["arm_empty", "B_on_A"], d_list=["holding_B","clear_A"], group="action")
        self.stack_B_on_C = Module("stack_B_on_C", c_list=["clear_C", "holding_B"], a_list=["arm_empty", "B_on_C"], d_list=["holding_B","clear_C"], group="action")
        self.stack_C_on_A = Module("stack_C_on_A", c_list=["clear_A", "holding_C"], a_list=["arm_empty", "C_on_A"], d_list=["holding_C","clear_A"], group="action")
        self.stack_C_on_B = Module("stack_C_on_B", c_list=["clear_B", "holding_C"], a_list=["arm_empty", "C_on_B"], d_list=["holding_C","clear_B"], group="action")
        #self.recognize_bottle = Module("recognize_bottle", c_list=["object_observed"], a_list=["bottle_observed"], d_list=["object_observed"], group="belief")
        #self.recognize_cup = Module("recognize_cup", c_list=["object_observed"], a_list=["cup_observed"], d_list=["object_observed"], group="belief")
        
        self.modules = [self.drink, self.bring_mouth_to_cup, self.bring_cup_to_mouth, self.put_bottle_down, self.pour_liquid, self.pick_up_bottle, self.put_cup_down, self.pick_up_cup]
    
    def make_dict(self):
        module_dict_action = {}
        module_dict_belief = {}
        for module in self.modules:
            if set(module.c_list).issubset(self.current_state):
                if module.group == "action":
                    module_dict_action[str(module.name)] = module.activation_level
                else:
                    module_dict_belief[str(module.name)] = module.activation_level
        return module_dict_action, module_dict_belief
    
    def external_spread(self):
        for module in self.modules:
            if "relieve_thirst" in module.a_list:
                module.activation_level += self.influence_from_goals # ①
            if "cup_observed" in module.c_list:
                module.activation_level += self.cup_intensity # ②
            if "bottle_observed" in module.c_list:
                module.activation_level += self.bottle_intensity
                
    def internal_spread(self):
        for module in self.modules:
            l = copy.copy(self.modules)
            l.remove(module)
            if set(module.c_list).issubset(self.current_state):
                for i in l:
                    if len(set(module.a_list) & set(i.c_list)) > 0:
                        i.activation_level += self.influence_from_state # ⑤
            else:
                for i in l:
                    if len(set(i.a_list) & set(module.c_list)) > 0:
                        i.activation_level += self.influence_from_state # ④
                
        
    def decay_function(self):
        for module in self.modules:
            if "be_polite" in module.d_list:
                module.activation_level -= self.influence_from_achieved_goals # ③
            l = copy.copy(self.modules)
            l.remove(module)
            for i in l:
                if len(set(module.c_list) & set(i.d_list)) > 0:
                    i.activation_level -= self.influence_from_state # ⑥
    
    def selection(self):
        action, belief = self.make_dict()
        keys_of_action_max = []
        keys_of_belief_max = []
        if len(action) != 0:
            action_max_val = max(action.values())
            keys_of_action_max = [key for key in action if action[key] == action_max_val]
        if len(belief) != 0:
            belief_max_val = max(belief.values())
            keys_of_belief_max = [key for key in belief if belief[key] == belief_max_val]

        return keys_of_action_max, keys_of_belief_max
    
    def global_loop(self):
        self.current_state = self.env.get_current_state()
        self.external_spread()
        self.internal_spread()
        self.decay_function()
        action, belief = self.selection()
        if len(action) == 1:# activation_levelの最大値をとるモジュールが2つ以上ある場合は無視する．
            for module in self.modules:
                if module.name == action[0]:
                    if module.act() == True:
                        self.env.do_action(module)
                        module.activation_level = 0
                        for state in module.a_list:
                            self.current_state.add(state)
                        for state in module.d_list:
                            self.current_state.discard(state)
"""                    
        if len(belief) == 1:
            for module in self.modules:
                if module.name == belief[0]:
                    if module.act() == True:
                        module.activation_level = 0
                        for state in module.add_list:
                            self.current_state.add(state)
                        for state in module.delete_list:
                            self.current_state.discard(state)
"""
```
